## 15 . try to find out the 10th most loss making product catagory.

```python
cursor.execute("USE liveclasstask31july")
```

```python
query = """select distinct o.SC,o.MP 
from (
		select `Product Category` AS C,
			   `Product Sub-Category` AS SC,
			    AVG(`Profit`) over(partition by `Product Sub-Category`) AS MP 
		from orders) as o 
order by o.MP limit 10
"""
query = query.replace('\n','').replace('\t','')
query
```

```python
cursor.execute(query)
dict(cursor.fetchall())
```
